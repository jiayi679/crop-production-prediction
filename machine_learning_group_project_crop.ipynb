```python
# Part 1: Data preprocessing (clean full_weather_cleaned dataset）
import pandas as pd
# Load the raw weather dataset
df6 = pd.read_csv('/content/full_weather_cleaned - full_weather_cleaned.csv')

# Display column names
print("\nColumn names:")
print(df6.columns)

# Check for duplicates
print("\nChecking for duplicates...")
duplicate_rows = df6[df6.duplicated()]
print(f"Duplicate rows found: {len(duplicate_rows)}")

# Remove duplicate rows
df6.drop_duplicates(inplace=True)

# Check for missing values before handling
print("\nMissing values per column before handling:")
print(df6.isnull().sum())

# Fill missing numeric values with median (or mean as alternative)
df6.fillna(df6.median(numeric_only=True), inplace=True)

# Check again for missing values after filling
print("\nMissing values per column after handling:")
print(df6.isnull().sum())

# Rename 'city' column to 'district' if present
if 'city' in df6.columns:
    df6 = df6.rename(columns={'city': 'district'})
    print("Column 'city' renamed to 'district'.")

# Part 1.2: Filter df6 using valid (state, district) from crop dataset
# Load crop production dataset
df1 = pd.read_csv('/content/cleaned_crops_district_production.csv')

# Strip spaces from 'state' and 'district'
df1['state'] = df1['state'].str.strip()
df1['district'] = df1['district'].str.strip()

# Get valid (state, district) combinations
valid_locations = df1[['state', 'district']].drop_duplicates().reset_index(drop=True)

# Clean spaces in df6 columns
df6['state'] = df6['state'].str.strip()
df6['district'] = df6['district'].str.strip()

# Inner join to keep only matching (state, district) rows
df6_filtered = df6.merge(valid_locations, on=['state', 'district'], how='inner')

# Part 1.3: Group by state and district, calculate averages
# Sort for readability
df6_sorted = df6_filtered.sort_values(by=['state', 'district']).reset_index(drop=True)

# Group by state and district, calculate mean for each numeric column
df_grouped = df6_sorted.groupby(['state', 'district'], as_index=False).mean()

# Preview and save the grouped result
print("\nGrouped by state and district (averaged values):")
print(df_grouped.head())
# Save only the final grouped result
df_grouped.to_csv('/content/filtered_grouped_avg.csv', index=False)
print("\nFinal grouped dataset saved as 'filtered_grouped_avg.csv'.")
```

```python
#Part1:Data preprocessing(merge all datasets together)
import pandas as pd
import matplotlib.pyplot as plt
# Load all cleaned datasets
df1 = pd.read_csv('/content/cleaned_crops_district_production.csv')  # Contains crop_species & crop_type
df2 = pd.read_csv('/content/cleaned_air_pollution_district.csv')  # Contains district & state, keep API only
df3 = pd.read_csv('/content/cleaned_gdp_district_avg.csv')  # Contains district & state, keep value only
df4 = pd.read_csv('/content/cleaned_malaysia_elevation.csv')  # Contains district & state, keep Average elevation(m)
df5 = pd.read_csv('/content/cleaned_crops_district_area.csv')  # Contains district & state, keep planted_area
df6 = pd.read_csv('/content/filtered_grouped_avg.csv')  # Contains state & district, keep weather columns

# Pre-aggregate df1 and df5 to prevent duplication
df1_grouped = df1.groupby(['state', 'district', 'crop_type', 'crop_species'], as_index=False)['production'].sum()
df5_grouped = df5.groupby(['state', 'district', 'crop_type', 'crop_species'], as_index=False)['planted_area'].sum()

# Merge crop data
crop_df = df1_grouped.merge(df5_grouped, on=['state', 'district', 'crop_type', 'crop_species'], how='inner')

# Aggregate GDP and Weather data
df3_grouped = df3.groupby(['state', 'district'], as_index=False)['value'].mean()
df6_grouped = df6.groupby(['state', 'district'], as_index=False)[[
    'temperature', 'dew_point', 'humidity', 'wind_speed',
    'uv_index', 'solar_radiation', 'precipitation_rate', 'precipitation_total'
]].mean()
# Reduce datasets to only relevant columns
df2_reduced = df2[['state', 'district', 'API']]
df4_reduced = df4[['state', 'district', 'Average elevation(m)']]

# Merge location-related datasets
location_df = df2_reduced.merge(df3_grouped, on=['state', 'district'], how='outer') \
                         .merge(df4_reduced, on=['state', 'district'], how='outer') \
                         .merge(df6_grouped, on=['state', 'district'], how='outer')
# Merge crop_df with location_df
final_df = crop_df.merge(location_df, on=['state', 'district'], how='inner')

# Handle missing values
print("\nMissing values before handling:")
print(final_df.isnull().sum())

# Drop rows with missing values
final_df.dropna(inplace=True)

print("\nMissing values after dropping rows with NA:")
print(final_df.isnull().sum())
# Save the final dataset
final_df.to_csv('/content/final_merged_dataset.csv', index=False)
print("\nPreprocessing complete! Final dataset saved as 'final_merged_dataset.csv'.")
print(final_df.head(2))
```

```python
#Part 2:feature selection
from sklearn.model_selection import train_test_split
# Unified Feature Selection by correlation threshold
# Calculate the absolute correlation of all numeric columns with the target variable 'production'
correlation = final_df.corr(numeric_only=True)['production'].abs()

# Set a correlation threshold
threshold = 0.015

# Select all features with correlation greater than the threshold, excluding the target variable itself
selected_features = correlation[correlation > threshold].drop('production').index.tolist()
print("Selected features based on correlation threshold:")
print(selected_features)

# Prepare the feature set for training, keep selected features + non-feature columns (state, district, crop_type, crop_species)
non_feature_cols = ['state', 'district', 'crop_type', 'crop_species']
X_all = final_df[non_feature_cols + selected_features]
y_all = final_df['production']
print("Number of selected features:", len(selected_features))
```

```python
#Part 3: Train-Test Split (8:2)
X_train, X_test, y_train, y_test = train_test_split(
    X_all, y_all, test_size=0.2, random_state=42
)
print("Final dataframe shape:", final_df.shape)
print(f"\n Dataset split complete!")
print(f"Training set size: {X_train.shape[0]} samples")
print(f"Testing set size : {X_test.shape[0]} samples")

# Check head rows
print("\n Training set preview:")
print(X_train.head())
```

```python
# Part 4: Machine Training (Linear Regression)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
# 1. One-hot encode the training and test datasets
X_train_encoded = pd.get_dummies(X_train)
X_test_encoded = pd.get_dummies(X_test)

# 2. Align test set columns with training set (to avoid missing dummy variables)
X_train_encoded, X_test_encoded = X_train_encoded.align(X_test_encoded, join='left', axis=1, fill_value=0)

# 3. Initialize the linear regression model
model = LinearRegression()

# 4. Fit the model
model.fit(X_train_encoded, y_train)

# 5. Make predictions
y_train_pred = model.predict(X_train_encoded)
y_test_pred = model.predict(X_test_encoded)

# 6. Evaluate performance
print("📊 Linear Regression Performance:")
print("Training set size:", X_train_encoded.shape[0])
print("Testing set size :", X_test_encoded.shape[0])
print("Train RMSE:", np.sqrt(mean_squared_error(y_train, y_train_pred)))
print("Test RMSE :", np.sqrt(mean_squared_error(y_test, y_test_pred)))
print("R² Score  :", r2_score(y_test, y_test_pred))

# Plot: True vs. Predicted values for training set
plt.figure(figsize=(12,5))
plt.subplot(1, 2, 1)
plt.scatter(y_train, y_train_pred, alpha=0.5, color='blue')
plt.plot([y_train.min(), y_train.max()], [y_train.min(), y_train.max()], 'r--', lw=2)  # Reference line y=x
plt.xlabel("True Values (Train)")
plt.ylabel("Predictions (Train)")
plt.title("Training Set: True vs Predicted")

# Plot: True vs. Predicted values for test set
plt.subplot(1, 2, 2)
plt.scatter(y_test, y_test_pred, alpha=0.5, color='green')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)  # Reference line y=x
plt.xlabel("True Values (Test)")
plt.ylabel("Predictions (Test)")
plt.title("Testing Set: True vs Predicted")
plt.tight_layout()
plt.show()
```

```python
# Part 4: Machine Training (Ridge Regression)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error, r2_score
# One-hot encode the training and test sets and align columns
X_train_encoded = pd.get_dummies(X_train)
X_test_encoded = pd.get_dummies(X_test)
X_train_encoded, X_test_encoded = X_train_encoded.align(X_test_encoded, join='left', axis=1, fill_value=0)

# Define a simple grid for Ridge alpha values
param_grid = {
    'alpha': [0.01, 0.1, 1.0, 10.0, 100.0]
}
# Initialize Ridge model
ridge = Ridge()

# Apply GridSearchCV for hyperparameter tuning
grid_search = GridSearchCV(ridge, param_grid, cv=3, scoring='neg_mean_squared_error', n_jobs=-1)
grid_search.fit(X_train_encoded, y_train)

# Get the best model from grid search
best_ridge = grid_search.best_estimator_

# Predict on train and test sets
y_train_pred = best_ridge.predict(X_train_encoded)
y_test_pred = best_ridge.predict(X_test_encoded)

# Evaluate performance
print("📊 Ridge Regression Performance:")
print("Training set size:", X_train_encoded.shape[0])
print("Testing set size :", X_test_encoded.shape[0])
print("Best alpha:", grid_search.best_params_['alpha'])
print("Train RMSE:", np.sqrt(mean_squared_error(y_train, y_train_pred)))
print("Test RMSE :", np.sqrt(mean_squared_error(y_test, y_test_pred)))
print("R² Score  :", r2_score(y_test, y_test_pred))

# Plot: True vs. Predicted values for training set
plt.figure(figsize=(12,5))
plt.subplot(1, 2, 1)
plt.scatter(y_train, y_train_pred, alpha=0.5, color='blue')
plt.plot([y_train.min(), y_train.max()], [y_train.min(), y_train.max()], 'r--', lw=2)
plt.xlabel("True Values (Train)")
plt.ylabel("Predictions (Train)")
plt.title("Training Set: True vs Predicted (Ridge)")

# Plot: True vs. Predicted values for test set
plt.subplot(1, 2, 2)
plt.scatter(y_test, y_test_pred, alpha=0.5, color='green')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)
plt.xlabel("True Values (Test)")
plt.ylabel("Predictions (Test)")
plt.title("Testing Set: True vs Predicted (Ridge)")
plt.tight_layout()
plt.show()
```

```python
# Part 4: Machine Learning Training (Random Forest Regression)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import GridSearchCV
# 1. One-hot encode training and test sets
X_train_encoded = pd.get_dummies(X_train)
X_test_encoded = pd.get_dummies(X_test)

# 2. Align the columns to make sure both sets have the same features
X_train_encoded, X_test_encoded = X_train_encoded.align(X_test_encoded, join='left', axis=1, fill_value=0)

# 3. Initialize base model
rf = RandomForestRegressor(random_state=42)

# 4. Define a small parameter grid for tuning (to keep training time reasonable)
param_grid = {
    'n_estimators': [100],          # number of trees in the forest
    'max_depth': [10, 20],          # maximum depth of each tree
    'min_samples_split': [2],       # minimum number of samples to split an internal node
    'min_samples_leaf': [1]         # minimum number of samples required to be a leaf node
}
# 5. Perform grid search with 3-fold cross-validation
grid_search = GridSearchCV(
    estimator=rf,
    param_grid=param_grid,
    cv=3,
    scoring='neg_mean_squared_error',
    n_jobs=-1,
    verbose=1
)
# 6. Fit the model with best hyperparameters
grid_search.fit(X_train_encoded, y_train)
best_model = grid_search.best_estimator_

# 7. Make predictions
y_train_pred = best_model.predict(X_train_encoded)
y_test_pred = best_model.predict(X_test_encoded)

# 8. Evaluate model performance
print("🌲 Random Forest Regression Performance:")
print("Training set size:", X_train_encoded.shape[0])
print("Testing set size :", X_test_encoded.shape[0])
print("Best parameters found:", grid_search.best_params_)
print("Train RMSE:", np.sqrt(mean_squared_error(y_train, y_train_pred)))
print("Test RMSE :", np.sqrt(mean_squared_error(y_test, y_test_pred)))
print("R² Score  :", r2_score(y_test, y_test_pred))

# 9. Visualization - Predicted vs. True values
plt.figure(figsize=(12, 5))

# Training set
plt.subplot(1, 2, 1)
plt.scatter(y_train, y_train_pred, alpha=0.5, color='blue')
plt.plot([y_train.min(), y_train.max()], [y_train.min(), y_train.max()], 'r--', lw=2)
plt.xlabel("True Values (Train)")
plt.ylabel("Predictions (Train)")
plt.title("Training Set: True vs Predicted")

# Testing set
plt.subplot(1, 2, 2)
plt.scatter(y_test, y_test_pred, alpha=0.5, color='green')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)
plt.xlabel("True Values (Test)")
plt.ylabel("Predictions (Test)")
plt.title("Testing Set: True vs Predicted")
plt.tight_layout()
plt.show()
```

```python
# Part 4: Machine Training (KNN)
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
# One-hot encode training and test sets, then align columns to avoid missing dummy columns in test set
X_train_encoded = pd.get_dummies(X_train)
X_test_encoded = pd.get_dummies(X_test)
X_train_encoded, X_test_encoded = X_train_encoded.align(X_test_encoded, join='left', axis=1, fill_value=0)

# Initialize KNN regressor
knn = KNeighborsRegressor(n_neighbors=3)

# Fit KNN model
knn.fit(X_train_encoded, y_train)

# Make predictions
y_train_pred = knn.predict(X_train_encoded)
y_test_pred = knn.predict(X_test_encoded)

# Evaluate model performance
print("📊 KNN Regression Performance:")
print("Training set size:", X_train_encoded.shape[0])
print("Testing set size :", X_test_encoded.shape[0])
print("Train RMSE:", np.sqrt(mean_squared_error(y_train, y_train_pred)))
print("Test RMSE :", np.sqrt(mean_squared_error(y_test, y_test_pred)))
print("R² Score  :", r2_score(y_test, y_test_pred))

# Plot: Predicted vs True values for training set
plt.figure(figsize=(12,5))
plt.subplot(1, 2, 1)
plt.scatter(y_train, y_train_pred, alpha=0.5, color='blue')
plt.plot([y_train.min(), y_train.max()], [y_train.min(), y_train.max()], 'r--', lw=2)  # Reference line y=x
plt.xlabel("True Values (Train)")
plt.ylabel("Predictions (Train)")
plt.title("Training Set: True vs Predicted")

# Plot: Predicted vs True values for test set
plt.subplot(1, 2, 2)
plt.scatter(y_test, y_test_pred, alpha=0.5, color='green')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)  # Reference line y=x
plt.xlabel("True Values (Test)")
plt.ylabel("Predictions (Test)")
plt.title("Testing Set: True vs Predicted")
plt.tight_layout()
plt.show()
```
